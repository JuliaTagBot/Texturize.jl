# TextureNet

We have two pre-trained models that apply two kinds of textures on our input images:
* Fire 
* Frost


```julia
using MXNet
using Texturize
using Interact
using Images
```

```julia
styles = ["normal", "fire", "frost"];
images = map(x -> string(x), collect(1:10));
```

Now, choose the input image and choose your style:

```julia
@manipulate for image in images, style in styles
    if style == "normal"
        load("input/$image.jpg")
    else
        texturize("input/$image.jpg", style, "style")
    end
end
```
