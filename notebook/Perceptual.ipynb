# Perceptual

We have pretrained models s0-s22


```julia
using Texturize
using Interact
using Images
```

```julia
styles = ["normal", "s4", "s6", "s8", "s10"];
images = map(x -> string(x), collect(1:10));
```

Now, choose the input image and choose your style:

```julia
@manipulate for image in images, style in styles
    if style == "normal"
        load("$(Pkg.dir("Texturize"))/input/$image.jpg")
    else
        perceptual("$(Pkg.dir("Texturize"))/input/$image.jpg", style)
    end
end
```
